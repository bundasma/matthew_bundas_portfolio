```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
train_set = pd.read_csv("training_set.csv")
train_meta = pd.read_csv("training_set_metadata.csv")
train_total = pd.merge(train_set,train_meta, on ="object_id")


test_set = pd.read_csv("test_set_batchs.csv")
test_meta = pd.read_csv("plasticc_test_metadata.csv")
test_total = pd.merge(test_set,test_meta, on ="object_id")
```

```python
def make_lightcurve(data,label,objid,filename):
    


    mjd = data["mjd"]

    band_0 = data["flux"][data["passband"] == 0]
    mjd0 = data["mjd"][data["passband"] == 0]

    band_1 = data["flux"][data["passband"] == 1]
    mjd1 = data["mjd"][data["passband"] == 1]

    band_2 = data["flux"][data["passband"] == 2]
    mjd2 = data["mjd"][data["passband"] == 2]

    band_3 = data["flux"][data["passband"] == 3]
    mjd3 = data["mjd"][data["passband"] == 3]

    band_4 = data["flux"][data["passband"] == 4]
    mjd4 = data["mjd"][data["passband"] == 4]

    band_5 = data["flux"][data["passband"] == 5]
    mjd5 = data["mjd"][data["passband"] == 5]



    plt.figure(figsize = (8,8))
    plt.scatter(mjd0,band_0,color = "purple", label = "u - 0",alpha = 0.5)
    plt.scatter(mjd1,band_1, color = "green", label = "g - 1",alpha = 0.5)
    plt.scatter(mjd2,band_2, color = "red", label = "r - 2",alpha = 0.5)
    plt.scatter(mjd3,band_3, color = "yellow", label = "i - 3",alpha = 0.5)
    plt.scatter(mjd4,band_4, color = "brown", label = "z - 4 ",alpha = 0.5)
    plt.scatter(mjd5,band_5, color = "black", label = "y - 5",alpha = 0.5)
    
    plt.title("Light curve of obj " + str(objid) + " class = " + str(label),fontsize = 20)
    plt.xlabel("mjd",fontsize = 16)
    plt.ylabel("flux",fontsize = 16)
    plt.tick_params(axis = "both", which = "major", labelsize = 14, length = 8, width = 2)
    
    plt.legend(bbox_to_anchor=(1, 1), loc='upper left', ncol=1,fontsize = 16)
    plt.tight_layout()
    
    plt.savefig("lightcurves/" + filename, dpi = 400)
```

```python
data615 = train_set[train_set["object_id"] == 615]
label615 = int(train_meta[train_meta["object_id"] == 615]["target"])
make_lightcurve(data615,label615,615,"lc_615.png")
```
